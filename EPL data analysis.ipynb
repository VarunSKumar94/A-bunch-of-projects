```python
# a function to create list consisting 0910, 1011, 1112 ... 1819 where num1 is '09' and num2 is '19'
def combine(num1, num2): 
    combine  = [''.join([str("{:02d}"                   # converting num1 to a string and using 2 integer places so forming season as 0910 rather than 1011
                             .format(num)), 
                         str("{:02d}".format(num+1))]) for num in range(num1,num2)]
    return combine
```

```python
combine(9,19)
```

```python
# concat all csvs for seasons between year 9 and 19 and assigning a new column seas

allyears = pd.concat(pd.read_csv(f'season-{seas}_csv.csv')
                     .assign(seas=seas)
                     for seas in combine(9,19))
```
